# Why use Git?

Version control systems start with a base version of the document and then record changes you make each step of the way. You can think of it as a recording of your progress, you can rewind to start at the base document and play back each change you made, eventually arriving at your more recent version.

Once you think of changes as separate from the document itself, you can then think about playing back different sets of changes on the base document, ultimately resulting in different versions of that document. For example, two users can make independent sets of changes on the same document.

# Git Conflicts

As soon as people can work in parallel, theyll likely step on each others toes. This will even happen with a single person: if we are working on a piece of software on both our laptop and a server in the lab, we could make different changes to each copy. Version control helps us manage these conflicts by giving us tools to resolve overlapping changes.

To see how we can resolve conflicts, we must first create one.
If the collaborator now adds another station to this file. WTES # Wellington
and commits those changes and pushes to the repository.

```python
!git add stations.txt
!git commit -m "Adding Wellington"
!git push origin main
```

Now the owner make a different change to their copy without updating from GitHub. By adding another station that isn't wellington. Say ROTS.

You can add and commit this change.

```python
!git add stations.txt
!git commit -m "Adding new station"
```

However you will not be able to push due to conflicting changes.

```python
!git push origin main
```

Git rejects the push because it detects that the remote repository has new updates that have not been incorporated into the local branch. What we have to do is pull the changes from GitHub, merge them into the copy were currently working in, and then push that.

To do this first we need to pull the changes from the remote repository.

```python
!git pull origin main
```

After the changes from remote branch have been fetched, Git detects that changes made to the local copy overlap with those made to the remote repository, and therefore refuses to merge the two versions to stop us from trampling on our previous work. The conflict is marked in in the affected file.

If we look at our file we shall see the following:

Our change is preceded by <<<<<<< HEAD. Git has then inserted ======= as a separator between the conflicting changes and marked the end of the content downloaded from GitHub with >>>>>>>. (The string of letters and digits after that marker identifies the commit weve just downloaded and will be different to above)

It is now up to us to edit this file to remove these markers and reconcile the changes. We can do anything we want, keep the change made in the local repository, keep the change made in the remote repository, write something new to replace both, or get rid of the change entirely. Lets say we realize that we just want CCCC and ROTS in the file and not WTES so we edit the file by removing the HEAD ==== and >>> and just leave the file how we want it to be.

```python
!git add stations.txt
!git status
```

```python
!git commit -m "Merge changes from GitHub"
!git push origin main
```

The collaborator does not need to deal with this conflict as it is now dealt with on the remote repository. Now they just need to do a pull and they will get the merged changes.

# Pull Requests and Merging

Pull Requests is a good practice for merging branches back into the master (or main) branch when working in collaboration with others. This allows for reviewing of the changes made by a person and offers an area for collaborators to give feedback on changes.

To create a pull request, go to your git repository online and any new changes recently made to branches will have a pop-up to auto create a pull request. You can also naviagte to this by going to Pull Requests and click "New pull request" and then select the branch your wanting to merge into.

When you have created a pull request you can specify a few options to help show collaborators what these new changes are adding.
- In the write section you can add a small description of the changes, add links if necessary etc.
- You can add reviewers to the right when they have been added to your git repository

Reviewers then have the oppurtunity to look through the Pull Request and see the changes in the Files Changed tab. They can decide to Approve these changes or leave a Review commenting at given sections of the code to start a discussion for potential changes or questions.

To write comments at given sections of the code you can click on a line and add a comment. This can be a single comment or if your going to write a few comments you can start a review and then submit all the comments at once when you finish your review with the review changes button.

After collaborators are happy you can then merge the changes into the master (or main branch) as long as their is no conflicts with the master (or main) branch.

If you are not colaborating then you can just merge branches into master (or main) by using the PyCharm helper in the bottom right. Ensure your on the main branch by using checkout at the top, then you can hit merge 'example' into 'master' to do the merge. Any conflicts or issues will be raised by PyCharm and you will be directed as to what you should do.

# Basic qcore demo

Now we need to install qcore to our Enviournment so that we can import and use it within PyCharm. In the PyCharm terminal run the folllwing command replacing the [repo_directory] with the path to the folder location for all gmsim repositories.

```python
pip install -e [repo_directory]/qcore
```

Now we will run the folllowing to demo one of the functions within qcore.

```python
from qcore import siteamp_models
import matplotlib.pyplot as plt

dt = 0.005
n = 10000
vref = 500
vs = 350
vpga = 500
pga = 1

siteamp_models.init_ba18()
ampfi = siteamp_models.ba18_amp(dt, n, vref, vs, vpga, pga)

print(ampfi)

vs2 = 500
ampfi2 = siteamp_models.ba18_amp(dt, n, vref, vs2, vpga, pga)

vs3 = 600
ampfi3 = siteamp_models.ba18_amp(dt, n, vref, vs3, vpga, pga)

plt.plot(ampfi, label=f'{vs}')
plt.plot(ampfi2, label=f'{vs2}')
plt.plot(ampfi3, label=f'{vs3}')
plt.legend()
plt.show()
```

```python
import numpy as np

from IM_calculation.IM.intensity_measures import get_spectral_acceleration, calculate_Nstep
from qcore.constants import Components

periods = np.asarray([0.25, 0.5, 1.0, 2.0, 4.0])
DT = 0.005
time_steps = np.arange(0, 2 + DT, DT) * np.pi
waveform = np.sin(time_steps)
NT = waveform.size
ims = ["pSA"]
comps_to_store = {Components.c090}
Nstep = calculate_Nstep(DT, NT)

res = get_spectral_acceleration(waveform, periods, NT, DT, Nstep).squeeze()
print(res)
max_v = np.max(np.abs(res), axis=1)
print(max_v)

import matplotlib.pyplot as plt

plt.plot(time_steps, waveform, label="Input")
for p, wave in zip(periods, res):
    plt.plot(time_steps, wave, label=f"Output pSA {p}s")
plt.legend()
plt.show()
```
